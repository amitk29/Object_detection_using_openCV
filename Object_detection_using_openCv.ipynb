```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Creating Classifier 

```python
face_classifier=cv.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
```

### Reading Images

```python
img=cv.imread('Images/eye_face.jpg')
img_c=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img_c)
```

# 1. Detect Faces

```python
faces=face_classifier.detectMultiScale(img,1.3,5)
```

### No face Detected 

```python
if faces is ():
    print('No faces detected')
```

```python
def detect_face(img):
    face_rects=face_classifier.detectMultiScale(img)
    
    for (x,y,w,h) in face_rects:
        cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),10)
    
    return img
```

```python
result=detect_face(img_c)
plt.imshow(result)
```

# 2. Detect Eyes

```python
img=cv.imread('Images/eye_face.jpg')
img_c=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img_c)
```

```python
eyes_classifier=cv.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
```

```python
def eye_detector(img):
    eyes_classifier=cv.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
    eyes=eyes_classifier.detectMultiScale(img)
    
    for (x,y,w,h) in eyes:
        cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),10)
        
    return img    
```

```python
result=eye_detector(img_c)
plt.imshow(result)
```

# 3. Pedestrian Detection

```python
pedes_classifier=cv.CascadeClassifier('HaarCascades/haarcascade_fullbody.xml')
```

```python
capture=cv.VideoCapture('Video/People_Walking.mp4')
```

```python
while True:
    isTrue,frame=capture.read()
    
    pedes=pedes_classifier.detectMultiScale(frame)
    
    if isTrue:
        for (x,y,w,h) in pedes:
            cv.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),10)
            cv.imshow('Pedestrians',frame)
        if cv.waitKey(1)==27 :
            break
    else:
        break

capture.release()
cv.destroyAllWindows()
```

#  4. Car Detector

```python
car_classifier=cv.CascadeClassifier('HaarCascades/haarcascade_car.xml')
```

```python
capture=cv.VideoCapture('Video/Vehicles.mp4')
```

```python
while True:
    isTrue,frame=capture.read()
    car=car_classifier.detectMultiScale(frame)
    
    if isTrue:
        for (x,y,w,h) in car:
            cv.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),8)
            cv.imshow('Video',frame)
        if cv.waitKey(1)==27:
            break
        
    else:
        break
        
capture.release()
cv.destroyAllWindows()
```

#  6. Car Plate Detector

```python
img=cv.imread('Images/car_plate2.jpg')
img_c=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img_c)
```

```python
carplate_cassifier=cv.CascadeClassifier('HaarCascades/haarcascade_plate_number.xml')
```

```python
def car_plate(img):
    carplate_cassifier=cv.CascadeClassifier('HaarCascades/haarcascade_plate_number.xml')
    car_plate=carplate_cassifier.detectMultiScale(img)
    
    for (x,y,w,h) in car_plate:
        cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),5)
    
    return img
    
```

```python
result=car_plate(img_c)
plt.imshow(result)
```
